## Medidas de posicion: Percentiles

- Dividir el conjunto de los datos

- Datos ordenados de menor a mayor

Los estadsticos que tienen los datos ordenados se denominan de orden (mediana)

Estadisticos de orden:
- Mediana
- Rango: Diferencia entre el valor mximo y mnimo que tiene un conjunto
- Percentiles / Cuartiles / Etc


El percentil es un estadstico de orden que divide al conjunto en 100 partes iguales. 



- Ventajas:
1. Robustas (menos sensibles a valores atpicos)

Rango intercuartil: Diferencia entre Q3 y Q1 = P75 - P25
Valor X > 1.5 (Q3 - Q1)
