### Pulsaciones  vs  Cansancio (0-100)



```python
x = np.array([50,61,63,78,80,100,110,120,170])
y = np.array([20,40,42,55,65,72,80,84,99])

plt.plot(x,y, c='green', label='Cansancio promedio')
plt.xlabel("Pulsaciones")
plt.ylabel("Cansancio")
plt.grid() # grilla en mi figura
plt.title("Pulsaciones vs Cansancio", loc='center') # 'best'
plt.legend(loc='best') # fran no se lo acordaba :P 
plt.show()
```

## Subplots

Con subplot() podemos dibujar mÃºltiples grÃ¡ficos en una sola figura
La funciÃ³n subplot toma 3 argumentos (posicionales) que describen un layout estÃ¡ organizando en filas y columnas

plt.subplot(1,2,1)

```python
x1 = np.array([0,1,2,3])
y1 = np.array([1,2,4,5])
x2 = np.array([0,1,2])
y2 = np.array([2,4,6])

plt.subplot(1,2,1)
# color | c enum
plt.plot(x1,y1,'#ffefd5',label='grafico papayawip') #primer columna de mi layout
plt.grid()
plt.title("Primer grÃ¡fico")
plt.legend(loc='best') # fran no se lo acordaba :P 
plt.subplot(1,2,2)

plt.plot(x2,y2,c='r', label='otro grÃ¡fico')
plt.legend(loc='best') # fran no se lo acordaba :P 
plt.grid()
plt.suptitle("El titulo del proyecto / todos los graficos")
plt.show()
```

Vamos a cruzar sepal_length vs sepal_width (x,y) vamos a empezar a mezclar este grÃ¡fico con el color

```python
set(data.species) # es un tipo de datos que no admite duplicados
```

```python
plt.scatter(data['sepal_length'],data['sepal_width'],c='r')
```

```python
colors = {
    'setosa':'red',
    'versicolor':'blue',
    'virginica':'pink'
}
```

```python
for specie in data.species:
    if specie === 'setosa':
        return 'red'
    elif specie=='versicolor':
        return 'blue'
    else:
        return 'pink'
```

```python
species_color = pd.Series(data.species.map(colors))
species_color
```

```python
data['species_color'] = species_color
```

```python
data.sample(2)
```

```python
plt.scatter(data.sepal_length,data.sepal_width,c=data.species_color)
plt.show()
```

# GrÃ¡fico de barras

Discretas (diagrama de frecuencias)
Frecuencia: relaciÃ³n que cuenta las apariciones de un elemento dentro de un conjunto

```python
gender = ['Hombre','Mujer','Otro']
quantity = [20,30,10]

plt.bar(gender,quantity,color=['red','green','blue'])
plt.xlabel("Genero")
plt.ylabel("Frecuencia")
plt.grid()
plt.show()
```
