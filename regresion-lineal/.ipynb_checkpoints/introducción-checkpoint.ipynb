$e_1 = \hat{y_1} - y_1$

$e_2 = \hat{y_2} - y_2$
...

$e_5 = \hat{y_5} - y_5$

$Error = \Sigma_{i=1}(\hat{y_i} - y_i) = e_1 + e_2 + e_3 + ... e_n$

## MAE (Mean Absolute Error)

$Error = \frac{\Sigma_{i=1}|\hat{y_i} - y_i|}{N}$

## MSE (Mean Squared Error)

$Error = \frac{\Sigma_{i=1}(\hat{y_i} - y_i)^2}{N}$

Minimizar esta función de coste o minimizar el error

## RMSE 

$RMSE = \sqrt{\frac{\Sigma_{i=1}(\hat{y_i} - y_i)^2}{N}}$

```python
X = X.reshape(-1,1) # -1 automágico
# -1: Calcular automaticamente el tamaño de manera que el vector X se ajuste al arreglo original
# 1: Vector columna: El arreglo resultante tiene que ser un vector columna
```

```python
model.fit(X,Y)
```
